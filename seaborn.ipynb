```python
#STRIP PLOT
# X AXIS= STRING, Y= NUMERICAL VALUES
#dodge=True(for male and female space or hue space)
#jitter=0.1 is used for space in plot

sns.stripplot(data=df,x="day", y="total_bill", hue= "sex",dodge=True,jitter=0.2)
plt.legend(bbox_to_anchor=(0.3,0,1,1))
plt.show()
```

```python
#STYLE AND COLOR PLOT
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("tips.csv")
df.head()

```

```python
sns.barplot(data=df,x="sex", y="tip")
sns.set_style(style="darkgrid")
plt.show()
```

```python
sns.barplot(data=df,x="sex", y="tip")
sns.set_style(style='ticks')
plt.show()
```

```python
#PALPLOT
sns.palplot(sns.color_palette("spring"))
plt.show()
```

```python
sns.palplot(sns.color_palette("viridis"))
plt.show()
```

```python
# MULTIPLE PLOT IN SEABORN(FACETGRID)

sns.FacetGrid(df, col="smoker",height=2)
plt.show()
```

```python
sns.FacetGrid(df, col="day")
plt.show()
```

```python
a=sns.FacetGrid(df, col="time",height=2,hue="sex")
a.map(sns.barplot, "day","tip")
plt.show()
```

```python
#RELATIONALNAL PLOT- between two data point (scatter plot by default)
# col="day"- on the basis of days

sns.relplot(data=df,x="tip",y="total_bill",hue="sex",kind="line",col="day")
plt.show()
```

```python
sns.relplot(data=df,x="tip",y="total_bill",hue="sex",col="day")
plt.show()
```

```python
#SWARM PLOT- not overlap to each other unlike strip plot
#IN swarm plot data doesn't overlap by default to each other unlike strip plot

sns.swarmplot(data=df, x="day", y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
sns.stripplot(data=df,x="day",y="total_bill",jitter=0.3,hue="sex")
plt.show()
```

```python
#KDE PLOT

sns.kdeplot(data=df,x="total_bill")
plt.show()
```
